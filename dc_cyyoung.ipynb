# Cleaning the Pitching Stats

```python
pitchers = pd.read_csv("pitchers.csv", encoding="utf-8")
```

```python
#Without this, can't find the names in the Player columns
pitchers['Name'] = pitchers['Name'].apply(lambda x: str(x).replace(u'\xa0', u' '))
```

```python
pitchers
```

```python
#Getting Rid of "*" and "#" at the end of the players' names
pitchers.rename(columns = {'Name':'Player'}, inplace=True)
pitchers["Player"] = pitchers["Player"].str.replace("*", "", regex=False)
pitchers["Player"] = pitchers["Player"].str.replace("#", "", regex=False)
```

```python
pitchers
```

```python
pitchers.columns
```

```python
#Removing unwanted columns
pitchers = pitchers.drop(pitchers.columns[[0,1]], axis=1)
pitchers = pitchers.drop(columns=['W-L%', 'ERA+'])
```

```python
pitchers.columns
```

```python
#Moving the Year Column
pitchers = pitchers[['Player', 'Age', 'Year', 'Tm', 'Lg', 'W', 'L', 'ERA', 'G', 'GS', 'GF', 'CG',
       'SHO', 'SV', 'IP', 'H', 'R', 'ER', 'HR', 'BB', 'IBB', 'SO', 'HBP', 'BK',
       'WP', 'BF', 'FIP', 'WHIP', 'H9', 'HR9', 'BB9', 'SO9', 'SO/W']]
```

```python
#Removing Sub-Headers
pitchers.drop(pitchers[pitchers['Player'] == 'Name' ].index, inplace = True)
pitchers.drop(pitchers[pitchers['Player'] == 'LgAvg per 180 IP' ].index, inplace = True)
```

```python
pitchers.tail()
```

```python
pitchers[pitchers['Player'] == 'Pedro Martinez']
```

```python
pitchers[pitchers['Player'] == 'A.J. Cole']
```

```python
#Function for Changing 'TOT' 
def single_row(df):
    if df.shape[0]==1:
        return df
    elif "TOT" in set(df["Tm"]):
        tm = df[df["Tm"] == "TOT"]
        tm1 = df[df["Tm"] != "TOT"]
        tm["Tm"] = tm1.iloc[-1,:]["Tm"]
        tm["Lg"] = tm1.iloc[-1,:]["Lg"]
        return tm
    else:
        return df


pitchers = pitchers.groupby(["Player", "Year", "Age"]).apply(single_row)
```

```python
#Dropping Index Levels
for i in range(0,3):
    pitchers.index = pitchers.index.droplevel()
```

# Cleaning Team Data

```python
teams = pd.read_csv("teams.csv", encoding="utf-8")
```

```python
teams
```

```python
del teams['Unnamed: 0']
```

```python
teams =  teams.replace('Los Angeles Angels of Anaheim', 'Los Angeles Angels')
```

```python
print(nl_combined['Tm'].unique())
print(len(nl_combined['Tm'].unique()))
print('---------------------')
print(al_combined['Tm'].unique())
print(len(al_combined['Tm'].unique()))
```

```python
print(teams['Tm'].unique())
print(len(teams['Tm'].unique()))
```

```python
#Over the years, teams change names or even cities
#Merging short hand city names with long form team names

nicknames = {}

with open("nicknames.csv", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
nl_combined["Tm"] = nl_combined["Tm"].map(nicknames)
al_combined["Tm"] = al_combined["Tm"].map(nicknames)
```
