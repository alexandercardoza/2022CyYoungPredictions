```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
years = list(range(1990,2022))
```

```python
url_start = 'https://www.baseball-reference.com/awards/awards_{}.shtml'
```

# Scraping Cy Young Data

```python
for year in years:
    url = url_start.format(year)
    response = requests.get(url)
    
    with open("cy_young/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(response.text)
```

```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome(executable_path="/Users/alexa/chromedriver")
```

```python
for year in years:
    url = url_start.format(year)
    response = requests.get(url)

    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)

    html = driver.page_source
    with open("cy_young/{}.html".format(year),"w+", encoding="utf-8") as f:
        f.write(html)
```

# Parsing Cy Young Data with BeautifulSoup

```python
al_cy = []

for year in years:
    with open("cy_young/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    #Isolating the table using its id
    al_table = soup.find_all(id="AL_CYA_voting")
    #[0] only takes the first element of a list of dfs
    al = pd.read_html(str(al_table))[0]
    #Create a column for the year
    al["Year"] = year
    
    al_cy.append(al)
```

```python
nl_cy = []

for year in years:
    with open("cy_young/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    #Isolating the table using its id
    nl_table = soup.find_all(id="NL_CYA_voting")
    #[0] only takes the first element of a list of dfs
    nl = pd.read_html(str(nl_table))[0]
    #Create a column for the year
    nl["Year"] = year
    
    nl_cy.append(nl)
```

```python
al_cy = pd.concat(al_cy)
nl_cy = pd.concat(nl_cy)
```

# Scraping Pitcher Stats with Selenium

```python
years = list(range(1990,2023))
```

```python
pitchers_url_start = 'https://www.baseball-reference.com/leagues/majors/{}-standard-pitching.shtml'
```

```python
# Getting the Response Text
for year in years:
    pitchers_url = pitchers_url_start.format(year)
    pitchers_response = requests.get(pitchers_url)
    
    with open("pitcher_stats/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(pitchers_response.text)
```

```python
# Scraping Data with Selenium
driver = webdriver.Chrome(executable_path="/Users/alexa/chromedriver")
for year in years:
    pitchers_url = pitchers_url_start.format(year)
    url = pitchers_url.format(year)
    
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)
    
    html = driver.page_source
    with open("pitcher_stats/{}.html".format(year),"w+", encoding="utf-8") as f:
        f.write(html)
```

# Parsing Pitcher Stats with BeautifulSoup

```python
pitcher_df = []

for year in years:
    with open("pitcher_stats/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    #Isolating the table using its id
    pitcher_table = soup.find_all(id="players_standard_pitching")
    #[0] only takes the first element of a list of dfs
    pitcher = pd.read_html(str(pitcher_table))[0]
    #Create a column for the year
    pitcher["Year"] = year
    
    pitcher_df.append(pitcher)
```

```python
pitchers = pd.concat(pitcher_df)
```

```python
pitchers.to_csv("pitchers.csv", encoding="utf-8")
```

# Downloading Team Data

```python
teams_url = 'https://www.baseball-reference.com/leagues/majors/{}-standings.shtml'
```

```python
for year in years:
    url = teams_url.format(year)

    data = requests.get(url)

    with open("team/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
```

# Parsing Team Data

```python
dfs = []
dfs_e = []
dfs_c = []
dfs_w = []
for year in years:
    with open("team/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()

    #East Divisions
    soup = BeautifulSoup(page, "html.parser")
    team_table = soup.find_all(id="div_standings_E")  
    team_al = pd.read_html(str(team_table))[0]
    team_al["Year"] = year
    team_al["Division"] = "AL East"
    dfs.append(team_al)
    team_nl = pd.read_html(str(team_table))[1]
    team_nl["Year"] = year
    team_nl["Division"] = "NL East"
    dfs.append(team_nl)


    #West Divisions
    soup = BeautifulSoup(page, "html.parser")
    team_table = soup.find_all(id="div_standings_W")
    team_al = pd.read_html(str(team_table))[0]
    team_al["Year"] = year
    team_al["Division"] = "AL West"
    dfs.append(team_al)
    team_nl = pd.read_html(str(team_table))[1]
    team_nl["Year"] = year
    team_nl["Division"] = "NL West"
    dfs.append(team_nl)


    #Central Divisions
    soup = BeautifulSoup(page, "html.parser")
    team_table = soup.find_all(id="div_standings_C")
    if len(team_table) == 0:
        continue
    else:
        team_al = pd.read_html(str(team_table))[0]
        team_al["Year"] = year
        team_al["Division"] = "AL Central"
        dfs.append(team_al)
        team_nl = pd.read_html(str(team_table))[1]
        team_nl["Year"] = year
        team_nl["Division"] = "NL Central"
        dfs.append(team_nl)
```
